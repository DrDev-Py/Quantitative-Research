* __Substituting the scoring system for the likert answers__

For questions where the correct option would be __True__ or __Good__ strongly agree would have a score of 4, while strongly disagree would have a score of 0

> i.e :

 > __Strongly Agree__ = 4
 
 > __Agree__ = 3
 
 > __Neutral__ = 2
 
 > __Disagree__ = 1
 
 > __Strongly Disagree__ = 0
 
 For questions where the correct option would be __False__ or __Bad__ strongly agree would have a score of 0, while strongly disagree would have a score of 4

> i.e :

 > __Strongly Agree__ = 0
 
 > __Agree__ = 1
 
 > __Neutral__ = 2
 
 > __Disagree__ = 3
 
 > __Strongly Disagree__ = 4

### Replacing columns Likert Scores

* __Replacing columns with correct options as good or true__

```python
# Creating the true/good column
true_good = [1, 4, 5, 6, 8, 10, 11, 14, 20,21,22,24,25,30]

# Transforming the column to zero indexing, so it matches with the dataframe
true_good2 = [x - 1 for x in true_good]

print(true_good2)
```

```python
type(microbes_likert)
```

```python
true_col_names = []
for num in true_good2: 
    true_col_names.append(microbes_likert.columns[num])
print(true_col_names)
print('\n',len(true_col_names))
```

```python
microbes = microbes.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)
microbes_likert = microbes_likert.applymap(lambda x: x.lower().strip() if isinstance(x, str) else x)
```

```python
replace_dict = {'strongly agree':4,  'agree':3, 'neutral':2, 'disagree':1, 'strongly disagree':0}
# microbes_likert[' [Antibiotics are useful for bacterial infections]']
```

```python
# mapping the columns to the dataframe as an access to substitute the likert scales
for col in true_col_names:
    microbes_likert.loc[:,col] = microbes_likert.loc[:,col].replace(replace_dict)

# # microbes_likert.loc[:,' [Antibiotics are useful for bacterial infections]']

# # col_names[0]
```

```python
false_col_names = [] # column names of false columns
for num in false_bad2: 
    false_col_names.append(microbes_likert.columns[num])
```

# Is there a correlation between the attitudes, opinions and knowledge about antimicrobial stewardship?

# Chi Square Test - Between Gender and Knowledge

```python
microbes.head()
```

# FOR KNOWLEDGE

```python
import pandas as pd
from scipy.stats import chi2_contingency

for i in microbes.columns[3:10]:
    

    # Create a contingency table from the dataframe
    contingency_table = pd.crosstab(microbes[i], microbes['knowledge'])

    # Perform the chi-square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    # Print the results
    print(f'FOR {i} \n \n')
    print("Chi-square statistic:", chi2)
    print("P-value:", p_value)
    print("Degrees of freedom:", dof)
    print("Expected frequencies:\n", expected)
    print("\n \n \n \n \n \n \n \n")
```
